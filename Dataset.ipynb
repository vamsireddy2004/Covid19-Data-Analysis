```python
sns.histplot(data['daily_vaccinations'], bins=30)
plt.title('Distribution of Daily Vaccinations')
plt.show()
```

```python
usa_data = data[data['country'] == 'United States']
print(usa_data.head())
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data
```

```python
usa_data['total_vaccinations_rolling'] = usa_data['daily_vaccinations'].cumsum()
print(usa_data[['date', 'daily_vaccinations', 'total_vaccinations_rolling']])
```

```python
country_vaccinations = data.groupby('country')['total_vaccinations'].max().reset_index()
print(country_vaccinations)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(usa_data['date'], usa_data['daily_vaccinations'], label='Daily Vaccinations')
plt.title('Daily COVID-19 Vaccinations in the US')
plt.xlabel('Date')
plt.ylabel('Number of Vaccinations')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
top_countries = country_vaccinations.nlargest(10, 'total_vaccinations')

plt.figure(figsize=(12, 6))
sns.barplot(x='total_vaccinations', y='country', data=top_countries)
plt.title('Top 10 Countries by Total Vaccinations')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
